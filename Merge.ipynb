```python
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
from sqlalchemy import extract
```

```python
import sqlite3
import pandas as pd
  
sql_list = []    

# Connect to SQLite database
conn = sqlite3.connect('Data/University.db')
  
# Load CSV data into Pandas DataFrame
stud_data = pd.read_csv('Data/oscar_filter.csv')
# Write the data to a sqlite table
stud_data.to_sql('oscar_filter', conn, if_exists='replace', index=False)
  
# Create a cursor object
cur = conn.cursor()
# Fetch and display result

# Close connection to SQLite database

stud_data = pd.read_csv('Data/OMDB API.csv')
# Write the data to a sqlite table
stud_data.to_sql('omdb_api', conn, if_exists='replace', index=False)
  
# Create a cursor object
cur = conn.cursor()
# Fetch and display result
for row in cur.execute('select oscar_filter.film, omdb_api.boxoffice, omdb_api.imdbrating, omdb_api.metascore, oscar_filter.year_film, oscar_filter.category, oscar_filter.winner from oscar_filter inner join omdb_api on oscar_filter.film = omdb_api.title'):
    sql_list.append(row)
    
conn.close()
    
```

```python
finale = pd.DataFrame(sql_list)
finale
```

```python
final = finale.rename(columns={0: "film", 1: "box office", 2: "imdb rating", 3: "metascore", 4: "year_film", 5: "category", 6: "winner"})
final
```

```python
final.winner = final.winner.astype(bool)
```
